#Retail Insights from Superstore Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv("Analysis of Super Store.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
print(df["Ship Mode"].unique())
print(df["Segment"].unique())
print(df["Country"].unique())
print(df["Category"].unique())
print(df["Sub-Category"].unique())
print(df["Region"].unique())
```

```python
df.groupby("Region")["Sales"].sum().plot.bar()
plt.title('Total Sales by Region')
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby("Region")["Profit"].sum().plot.bar()
plt.title('Total Profit by Region')
plt.ylabel('Profit')
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot.pie(autopct="%1.0f%%")
plt.title('Sales Distribution by Region')
plt.ylabel('')
plt.show()
```

```python
df.groupby("Region")["Profit"].sum().plot.pie(autopct="%1.0f%%")
plt.title('Profit Distribution by Region')
plt.ylabel('')
plt.show()
```

```python
df.groupby("Segment")["Sales"].sum().plot.bar()
plt.title('Total Sales by Segment')
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby("Category")["Profit"].sum().plot.pie(autopct="%1.0f%%")
plt.title('Profit Distribution by Category')
plt.ylabel('')
plt.show()
```

```python
df.groupby("State")["Sales"].sum().plot.bar()
plt.title('Total Sales by State')
plt.ylabel('Sales')
plt.show()
```

```python
df.groupby("State")["Profit"].sum().plot.bar()
plt.title('Total Profit by State')
plt.ylabel('Profit')
plt.show()
```
