```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/employees.csv")
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
data["SALARY"] = data["SALARY"].astype("float")
```

```python
package =data.groupby("EMPLOYEE_ID")["SALARY"].mean()
package.plot(kind = "bar")
for i in range(len(package.index)):
  plt.text(i, package[package.index[i]], str( round(package[package.index[i]],1)))
```

```python
data["SALARY"].value_counts().plot(kind="bar")
```

```python
for column in data.columns[1:]:
  if data[column].dtype != "object":
    sns.histplot(data[column],kde = True)
    plt.title(column)
    plt.show()
```

```python
data[(data["SALARY"]>0) & (data["SALARY"]<200000) ]["SALARY"].plot(kind = "hist")
```

```python
data[(data["SALARY"]>0) & (data["SALARY"]<17000) ]["SALARY"].plot(kind = "box")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['SALARY'], bins=20, kde=True)
plt.title('SALARY_DISTRIBUTION')
plt.xlabel('SALARY')
plt.ylabel('FREQUENCY')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='JOB_ID')
plt.title('JOB_ID_COUNTS')
plt.xlabel('Job_ID')
plt.ylabel('COUNT')
plt.xticks(rotation=45)
plt.show()
```
